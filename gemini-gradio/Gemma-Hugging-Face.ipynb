````python
tokenizer = AutoTokenizer.from_pretrained("google/gemma-2b-it")
````

**Download your model from Hugging Face**

````python
model = AutoModelForCausalLM.from_pretrained(
    "google/gemma-2b-it",
    torch_dtype=torch.bfloat16,
    device_map="auto"
)
````

````python
input_text = "Write 20 things you know about Australia"
````

**Convert input text to tokens**

````python
input_ids = tokenizer(input_text, return_tensors="pt").to("cuda")
````

````python
response = model.generate(**input_ids, max_new_tokens=512)
````

````python
response
````

**Convert the response into a proper Natural Language Text**

````python
output = tokenizer.decode(response[0])
print(output)
````

**Chat template**
* Let's use a chat template and pass it to the model and get an output for it.

````python
chat = [
    {"role": "user", "content": "Write a Python program to train a logistic regression model"}
]

prompt = tokenizer.apply_chat_template(chat, tokenize=False, add_generation_prompt=True)

input_ids = tokenizer.encode(prompt, add_special_tokens=False, return_tensors="pt").to("cuda")
response = model.generate(input_ids=input_ids, max_new_tokens=512)

output = tokenizer.decode(response[0])
print(output)
````

````python
import re
````

````python
start_pattern = r"```python"

end_pattern = r"```"

pattern = re.compile(rf"{start_pattern}.(.*?){end_pattern}", re.DOTALL)

match = pattern.search(output)

if match:
  python_code = match.group(1).strip()
  print(python_code)
else:
  print("No match found")
````
